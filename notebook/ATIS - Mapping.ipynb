```python
def get_unique_tag(filename):
    atis_tag = set()
    with open("/Users/slouvan/sandbox/cross-domain/data/atis-2.train.conll","r") as f:
        for line in f:
            fields = line.strip().split("\t")
            if len(fields) > 1:
                # collect the tag
                tag = fields[1]
                if tag.startswith("B-"):
                    atis_tag.add(tag.split("B-")[1])
                elif tag.startswith("I-"):
                    atis_tag.add(tag.split("I-")[1])

    return atis_tag

atis_tag_train = get_unique_tag("/Users/slouvan/sandbox/cross-domain/data/atis-2.train.conll") 
atis_tag_valid = get_unique_tag("/Users/slouvan/sandbox/cross-domain/data/atis-2.dev.conll") 
atis_tag_test = get_unique_tag("/Users/slouvan/sandbox/cross-domain/data/atis.test.iob.conll") 

atis_tag = set()
atis_tag.update(atis_tag_train)
atis_tag.update(atis_tag_valid)
atis_tag.update(atis_tag_test)

for el in atis_tag:
    print(el)
```

```python
from collections import defaultdict
atis2conll = defaultdict(str)
atis2conll['O'] = 'O'
with open("atis_conll_mapping.csv","r") as f:
    for line in f:
        fields = line.split(",")
        #print(fields[1].strip())
        atis2conll[fields[0]] = fields[1].strip()
atis2conll
```

```python
with open("/Users/slouvan/sandbox/cross-domain/data/atis.test.iob.conll","r") as f:
    with open("/Users/slouvan/sandbox/cross-domain/data/atis.test.iob.retagged_ner.conll","w") as g:
        for line in f :
            if len(line.split("\t")) > 1:
                word, atis_tag = line.strip().split("\t")
                prefix = ""
                if atis_tag.startswith("B-") :
                    prefix = "B-"
                    tag = atis_tag.split("B-")[1]
                elif atis_tag.startswith("I-"):
                    prefix = "I-"
                    tag = atis_tag.split("I-")[1]
                else :
                    tag = atis_tag
                if atis2conll[tag] != "O":
                    print("{}\t{} -> {}\t{}".format(word, atis_tag, word, prefix+atis2conll[tag]))
                    g.write("{}\t{}\n".format(word, prefix+atis2conll[tag]))
                else:
                    print("{}\t{} -> {}\t{}".format(word, atis_tag, word, atis2conll[tag]))
                    g.write("{}\t{}\n".format(word, atis2conll[tag]))
            else :
                print(line)
                g.write(line)
        
```
