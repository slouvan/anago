```python
with open(TRAINING_DATA_NER_PATH, "w") as f:
    for id_sent, sent in enumerate(x_train):
        #print("SENTENCE :"+str(sent))
        predictions = model.analyze(sent)

        for id_token, token in enumerate(sent):
            #print("{} {} {}".format(sent[id_token], y_train[id_sent][id_token], predictions[1][id_token]))
            f.write(sent[id_token]+"\t"+predictions[1][id_token]+"\t"+y_train[id_sent][id_token]+"\n")
        f.write("\n")


with open(DEV_DATA_NER_PATH, "w") as f:
    for id_sent, sent in enumerate(x_dev):
        predictions = model.analyze(sent)
        for id_token, token in enumerate(sent):
            #print("{} {} {}".format(sent[id_token], y_dev[id_sent][id_token], predictions[1][id_token]))
            f.write(sent[id_token]+"\t"+predictions[1][id_token]+"\t"+y_dev[id_sent][id_token]+"\n")
        f.write("\n")
        
with open(TEST_DATA_NER_PATH, "w") as f:
    for id_sent, sent in enumerate(x_test):
        predictions = model.analyze(sent)
        for id_token, token in enumerate(sent):
            #print("{} {} {}".format(sent[id_token], y_test[id_sent][id_token], predictions[1][id_token]))
            f.write(sent[id_token]+"\t"+predictions[1][id_token]+"\t"+y_test[id_sent][id_token]+"\n")
        f.write("\n")
```
